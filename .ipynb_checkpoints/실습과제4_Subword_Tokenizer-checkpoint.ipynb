# 14. Subword Tokenizer


## 1번문제  
- 7쪽에서 생성된 naver_review.txt 파일을 읽어서 3쪽에 있는 데이터들을 계산. 단어의 수는 공백으로 분리된 어절의 수를 의미


```python
import pandas as pd 
import sentencepiece as spm 
import urllib.request 
import csv

urllib.request.urlretrieve("https://raw.githubusercontent.com/e9t/nsmc/master/ratings.txt", filename="ratings.txt") 

naver_df = pd.read_table('ratings.txt')

naver_df = naver_df.dropna(how = 'any') # Null 값이 존재하는 행 제거 총 199992평이 존재 

# 결과를 naver_review.txt 파일에 저장
with open('naver_review.txt', 'w', encoding='utf8') as f:
    f.write('\n'.join(naver_df['document']))
```

- 생성된 txt 파일


### 1번 문제 풀이 설명

  - 문장 수 세기: 문장을 마치는 .단위로 구분하여 주었다. 
  - 문제 단어 수 세기: 문제 조건대로 어절, 즉 띄어쓰기 단위로 구분해주었다. (tokenize사용)
  - 고유 단어 수 : 띄어쓰기 언어가 저장된 counter을 통해 구해주는데, 단어별 갯수가 아닌, 사전에 해당하는 길이를 출력하도록 하여 고유 단어수를 파악

```python
# 1번문제 풀이 코드 
# 1번문제는 띄어쓰기(공백)을 단위로 단어의 갯수를 파악하고, 문장의 마침표 단위로 문장의 길이를 구해주었음
from nltk import word_tokenize
from collections import Counter

# 텍스트의 문장 수, 단어 수, 고유 단어 수 만을 활용할 수 있도록 다른 변수에 저장
file = open("naver_review.txt", encoding='UTF-8')
data = file.read()
# 리뷰 갯수
num_review = len(naver_df)

# 문장 갯수
num_sentence = len(data.split('.'))

words = word_tokenize(data)
vocab = Counter(words)

# 단어 수 
num_word = len(words)

# 고유 단어 수 
num_vocab = len(vocab)

desc_dict = {'구분': ['네이버 영화 리뷰'],'리뷰 갯수':[num_review],'문장 갯수':[num_sentence],'단어 수':[num_word], '고유 단어 수':[num_vocab]}
desc_table = pd.DataFrame(desc_dict)
desc_table
```

## 2번문제 
- 단어수를 10,000개와 20,000개로 늘려 8쪽의 sentencepiece를 실행하고 80001~80010번의 평에 대한 결과를 얻음

### 10000개의 서브워드 집합 &  80001~80010번 평에 대한 결과 출력


```python
#sentence piece 실행
spm.SentencePieceTrainer.Train('--input=naver_review.txt --model_prefix=naver --vocab_size=10000 --model_type=bpe --max_sentence_length=9999')
```

- 만들어진 단어장 모습(naver.vocab)


```python
# naver.model을 이용하여 분석
sp = spm.SentencePieceProcessor() 
vocab_file = "naver.model" 
sp.load(vocab_file)

# 80001 ~ 800010까지 저장
lines = naver_df[80000:80010]
lines = lines['document']
result_10000=[]

for line in lines:
    print(line)
    print(sp.encode_as_pieces(line))
    print(sp.encode_as_ids(line))
    result_10000.append([sp.encode_as_pieces(line),sp.encode_as_ids(line)])
    print()
```

### 20000개의 서브 워드 집합 & 80001~80010번 평에 대한 결과 출력


```python
spm.SentencePieceTrainer.Train('--input=naver_review.txt --model_prefix=naver --vocab_size=20000 --model_type=bpe --max_sentence_length=9999')
```

- 만들어진 단어장 모습 


```python
# naver.model을 이용하여 분석
sp = spm.SentencePieceProcessor() 
vocab_file = "naver.model" 
sp.load(vocab_file)


# 80001 ~ 800010까지 저장
lines = naver_df[80000:80010]
lines = lines['document']
result_20000=[]

for line in lines:
    print(line)
    print(sp.encode_as_pieces(line))
    print(sp.encode_as_ids(line))
    result_20000.append([sp.encode_as_pieces(line),sp.encode_as_ids(line)])
    print()
```

## 3번 문제
- naver_review.txt 파일 내용에 대해 Okt 형태소 분석기를 실행시킴. 결과에서 나타난 고유단어수를 계산. 80001~80010번의 평에 대한 결과에 대해 Okt를 실행하고 위 2번의 결과와 비교함.


```python
from konlpy.tag import Okt

# naver_review.txt형태소 분석 진행(morphs함수 이용)
okt = Okt()
okt_words = okt.morphs(data)
okt_vocab=Counter(okt_words)

print("고유단어 수:", len(okt_vocab))

# 80001 ~ 80010
lines = naver_df[80000:80010]
lines = lines['document']
result_okt = []

for line in lines:
    print(line)
    words = okt.morphs(line)
    print(words)
    result_okt.append(words)
    print()
```

- 10000개의 Subword Tokenizer&20000 Subword Tokenizer&Okt로 분석한 80001~80010의 결과 비교 
  - 20000개보다 10000개의 단어장에서 조사가 잘 구분되어 있었다 ( 예: _조사, 의 와 _조사의) 10000단어의 단어장이 조금 더 단어를 간결하게 나누는 느낌이었다. 그리고 형태소분석기(Okt)와 비교에 보았을 때 몇개의 불용언이 섞여있는 것을 제외하면 유사하게 단어를 끊어 분석하는 것을 알 수 있었다.

```python
lines = naver_df[80000:80010]
lines = list(lines['document'])

for i in range(len(lines)):
    print(str(80001+i)+"번째 문장: "+lines[i])
    print("10000개의 Subword Tokenizer 분석 결과:",result_10000[i][0])
    print("20000개의 Subword Tokenizer 분석 결과:",result_20000[i][0])
    print("Okt 형태소 분석기 분석 결과:",result_okt[i])
    print()
```
